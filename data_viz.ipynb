# Visualización de datos con Python

```python
import pandas as pd
import matplotlib.pyplot as plt

url = 'https://raw.githubusercontent.com/alura-es-cursos/2063-visualizacion-de-datos-con-bibliotecas-de-python/refs/heads/main/inmigrantes_canada.csv'

df = pd.read_csv(url)
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.set_index('Pais', inplace=True)
df
```

```python
anos = list(map(str, range(1980, 2014)))
anos
```

```python
colombia = df.loc['Colombia', anos]
colombia
```

```python
col_dic = {'Año': colombia.index.tolist(),
            'Inmigrantes': colombia.values.tolist()
           }
col_dic
```

```python
datos_col = pd.DataFrame(col_dic)
datos_col.tail()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 4))
plt.plot('Año', 'Inmigrantes', data=datos_col)
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010', '2014'])
plt.title('Inmigrantes de Colombia')
plt.xlabel('Años')
plt.ylabel('Inmigrantes')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(datos_col['Año'], datos_col['Inmigrantes'], data=datos_col)
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.set_title('Inmigrantes de Colombia')
ax.set_xlabel('Años')
ax.set_ylabel('Inmigrantes')
plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 4))

axs[0].plot(datos_col['Año'], datos_col['Inmigrantes'], data=datos_col)
axs[0].set_title('Inmigrantes de Colombia')
axs[0].set_xlabel('Años')
axs[0].set_ylabel('# Inmigrantes')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid(True)

axs[1].boxplot(datos_col['Inmigrantes'], data=datos_col)
axs[1].set_title('Boxplot de la inmigración de columbianos\nhacia Canada')
axs[1].set_xlabel('Colombia')
axs[1].set_ylabel('# Inmigrantes')
axs[1].grid(True)

plt.show()
```

```python
fig, axs = plt.subplots(2, 2 , figsize=(10, 6))
fig.subplots_adjust(hspace=0.5, wspace=0.3)
fig.suptitle('Inmigracion de 4 paises latinoamericanos a Canadá')

axs[0,0].plot(df.loc['Colombia', anos], label='Colombia')
axs[0,0].set_title('Colombia')


axs[0,1].plot(df.loc['Argentina', anos], label='Argentina')
axs[0,1].set_title('Argentina')

axs[1,0].plot(df.loc['Brasil', anos], label='Brasil')
axs[1,0].set_title('Brasil')

axs[1,1].plot(df.loc['Perú', anos], label='Perú')
axs[1,1].set_title('Perú')

ymin = 0
ymax = 7000

for ax in axs.ravel():
  ax.set_ylim(ymin, ymax)

for ax in axs.flat:
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))
    ax.set_xlabel('Años')
    ax.set_ylabel('# Inmigrantes')
    ax.grid()

plt.show()
```
